```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the dataset
file_path = "E:\Jupyter\Masai BA\SAL_BW_project_3\IMDB_TOP250.csv"
df = pd.read_csv(file_path)
```

```python
#Display first few rows
df.head()
```

```python
#Basic Statistics
print("\nDataset Info:")
df.info()
```

```python
print("\nSummary Statistics:")
df.describe()
```

```python
#Count number of movies
print(f"\nTotal Movies: {df.shape[0]}")
```

```python
#Missing values check
print("\nMissing Values Count:")
df.isnull().sum()
```

```python
#Bar Chart: Number of Movies per Genre
genre_counts = pd.concat([df["Genre 1"], df["Genre 2"], df["Genre 3"]]).value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Number of Movies per Genre")
plt.show()
```

```python
#Scatter Plot: IMDB Rating vs. Release Year
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Year", y="Rating", alpha=0.7, edgecolors="k")
plt.xlabel("Release Year")
plt.ylabel("IMDB Rating")
plt.title("IMDB Rating vs. Release Year")
plt.show()
```

```python
#Word Cloud: Common Words in Movie Titles
title_text = " ".join(df["Title"])
wordcloud = WordCloud(width=800, height=400, background_color="black", colormap="viridis").generate(title_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Movie Titles")
plt.show()
```

```python
#Histogram: IMDB Ratings Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["Rating"], bins=10, kde=True, color="blue", edgecolor="black")
plt.xlabel("IMDB Rating")
plt.ylabel("Frequency")
plt.title("Distribution of IMDB Ratings")
plt.show()
```

```python
#Top 10 Directors with Most Movies
top_directors = df["Director"].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="magma")
plt.xlabel("Number of Movies")
plt.ylabel("Director")
plt.title("Top 10 Directors with the Most Movies")
plt.show()
```

```python
#Convert Duration to Minutes
def convert_duration(duration):
    parts = duration.split()
    minutes = 0
    for part in parts:
        if "h" in part:
            minutes += int(part.replace("h", "")) * 60
        elif "m" in part:
            minutes += int(part.replace("m", ""))
    return minutes

df["Duration_Minutes"] = df["Duration"].apply(convert_duration)
print(df.head())
#Movie Duration Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["Duration_Minutes"], bins=15, kde=True, color="purple", edgecolor="black")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Durations")
plt.show()
```

```python
#Average IMDB Rating per Genre
genre_avg_rating = df.melt(id_vars=["Rating"], value_vars=["Genre 1", "Genre 2", "Genre 3"], 
                           var_name="Genre_Type", value_name="Genre").dropna()
genre_avg_rating = genre_avg_rating.groupby("Genre")["Rating"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=genre_avg_rating.values, y=genre_avg_rating.index, palette="coolwarm")
plt.xlabel("Average IMDB Rating")
plt.ylabel("Genre")
plt.title("Average IMDB Rating per Genre")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[["Year", "Rating", "Duration_Minutes", "Number of Reviews(Votes)"]].corr(), 
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
#Trends in Movie Ratings Over Time
df["Decade"] = (df["Year"] // 10) * 10
decade_avg_rating = df.groupby("Decade")["Rating"].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=decade_avg_rating.index, y=decade_avg_rating.values, marker="o", color="red")
plt.xlabel("Decade")
plt.ylabel("Average IMDB Rating")
plt.title("Trends in Average IMDB Rating Over Time")
plt.show()
```
